**Data Preprocessing**

**Define Target variable y and Feature x**

```python
mile.columns
```

```python
y = mile['mpg']
```

```python
y.shape
```

```python
x = mile[['displacement','horsepower','weight','acceleration']]
```

```python
x.shape
```

```python
x
```

**Scaling Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

```python
pd.DataFrame(x).describe()
```


**Train Test Split Data**



```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7, random_state=2529)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage = 23.4-1.05Dispalcement-1.68Horsepower-4.10Weight-0.115Acceleration+error**

**Predict Test Data**

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

**Explaination :**

This dataset is a collection of car attributes from various makes and models, primarily from the early 1970s. It contains information about the fuel efficiency, engine specifications, and other key characteristics of cars. Each entry in the dataset represents a particular car model and includes attributes like miles per gallon (mpg), the number of cylinders in the engine, engine displacement, horsepower, the car's weight, acceleration capabilities, the model year, the origin of the car, and the car's name.

The `mpg` (miles per gallon) column is a crucial measure of a vehicle's fuel efficiency, indicating how far the car can travel on a gallon of fuel. This is a critical metric for understanding the energy efficiency of different cars. The `cylinders` column reflects the number of cylinders in the engine, which correlates with the engine's power and fuel consumption—cars with more cylinders typically have more powerful engines but lower fuel efficiency. `Displacement`, measured in cubic inches, provides insight into the engine's size and overall capacity, while `horsepower` indicates the engine's power output, affecting the car's speed and acceleration.

Additional columns like `weight` and `acceleration` offer more information about the car's physical and performance characteristics. Heavier cars generally have lower fuel efficiency and require more power to maintain performance, while acceleration measures the time it takes for a car to reach 60 miles per hour from a standstill, indicating its quickness. The `model_year` column gives the manufacturing year of the car, which can be crucial for understanding trends in automotive design and technology over time.

The `origin` column categorizes the cars by their region of manufacture, such as the USA, Europe, or Japan, allowing for comparisons between vehicles produced in different parts of the world. Finally, the `name` column provides a label or identifier for each car model, typically including the manufacturer's name and the specific model designation.

This dataset is rich with potential for various types of analysis. Researchers and analysts can explore relationships between engine size, weight, and fuel efficiency, or they can track how car designs and performance metrics have evolved over time. Additionally, the dataset is useful for comparing automotive trends across different regions, offering insights into how different markets prioritize aspects like power, efficiency, and speed. Overall, this dataset is a valuable resource for studying the evolution of car technology and the trade-offs between performance and efficiency in vehicle design.
